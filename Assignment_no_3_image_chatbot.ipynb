```python
from google.colab import userdata
GEMINI_API_KEY  = userdata.get('GOOGLE_API_KEY_2')
```

```python
import google.generativeai as genai
```

```python
genai.configure(api_key=GEMINI_API_KEY)
```

```python
model = genai.GenerativeModel("gemini-1.5-flash")
```

```python
import requests
from PIL import Image
from io import BytesIO
from IPython.display import display

# Replace with the URL of your image
url = 'https://images.pexels.com/photos/3680219/pexels-photo-3680219.jpeg?auto=compress&cs=tinysrgb&dpr=1&w=500'

# Fetch the image from the URL
response = requests.get(url)

# Open the image using PIL
img = Image.open(BytesIO(response.content))

# Display the image inline in Jupyter Notebook
display(img)
```

```python
# Write a prompt and store it in some variable
prompt = """Describe the image in detail"""

```

```python
response = model.generate_content([prompt, img])
print(response.text)
```
